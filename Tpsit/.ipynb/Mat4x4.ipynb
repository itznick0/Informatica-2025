```python
import numpy as np
import matplotlib.pyplot as plt
matrice=np.zeros((4,4),dtype=int)
mat=matrice+150
print(mat)
plt.imshow(matrice, cmap='gray', vmin=0, vmax=255)
```

```python
matrice1=np.full((4,4),151)
print(matrice1)
plt.imshow(matrice1, cmap='gray', vmin=0, vmax=255)
```

```python
matrice2=np.full((4,4), 0)
matrice2[1:3,1:3]=1
print(matrice2)
plt.imshow(matrice2, cmap='gray', vmin=0, vmax=255)
```

```python
matrice3=np.full((4,4), 255)
matrice3[::3,::3]=200
plt.imshow(matrice3, cmap='gray', vmin=0, vmax=255)
```

```python
matrice4=np.where(matrice2==1,matrice3,matrice1)
print(matrice4)
plt.imshow(matrice4, cmap='gray', vmin=0, vmax=255)
```
